## Exercise 2 - Filtering

Return the following subsets of the dataframe.

1. How many listings are there with a price less than 100? 


2. Find how many listings there are in just Brooklyn.


3. Find how many listings there are in Brooklyn with a price less than 100.


4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.


5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  The prices for all other listings should be the same as the were before. 


6. What % of the rooms are private, and what % of the rooms are shared.  
    * Hint, use `.value_counts()`


```python
# 1. How many listings are there with a price less than 100? 

# spits number of liststings with a price less than 100
len(df[df['price'] < 100])
```

```python
# 2. Make a new DataFrame of listings in Brooklyn named `df_bk` 
# and find how many listings in just Brooklyn.

# df['neighbourhood_group'].unique()
# Output : ['Manhattan', 'Brooklyn', 'Queens', 'Staten Island', 'Bronx'] ** checking groups **

df_bk = df[df['neighbourhood_group'] == 'Brooklyn']
len(df_bk) 

# 18632 Listing in Brooklyn
```

```python
# 3. Find how many listings there are in Brooklyn with a price less than 100.

# Put each of these conditions in paretheses (cond 1) & (cond 2) 
# pandas cant do cond 1 & cond 2
len(df[(df['neighbourhood_group'] == 'Brooklyn') & (df['price'] < 100)])

# Ouput: 10473 Listings

# Alternatively we can use the other df we created and just deal with the price condition
# it still gives us the same answer
# len(df_bk[df_bk['price'] < 100])

# Ouput: 10473 Listings
```

```python
# 4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.
# How many total are there that have those names

host_names = ['Michael', 'David', 'John', 'Daniel']
check_df = df[df['host_name'].isin(host_names)]
len(check_df)
# OUTPUT: 1258 have those names

# check_df['host_name'].unique()
# CHECK IF OUTPUT IS CORRECT:
# OUTPUT: ['Daniel', 'David', 'Michael', 'John'] 
```

```python
# 5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  
# The prices for all other listings should be the same as the were before. 

df = pd.read_csv('data/listings.csv')
First_Five = list(df[df['neighbourhood'] == 'Williamsburg']['price'].head(5))
# First_Five
# Ouput: [109, 299, 131, 175, 80]

# Create a column using np.where **** SUPER HELPFUL FUNCTION *****
# if it doesn't meet the condition just leave it as is
df['adjusted_price'] = np.where(df['neighbourhood'] == 'Williamsburg', df['price'] + 100, df['price'] * 1)

# Also im just converting the first 5 rows of neighborhoods that are a part of williamsburg and returning
# only the adjusted price column
New_Five = list(df[df['neighbourhood'] == 'Williamsburg']['adjusted_price'].head(5))
# New_Five
# Output: [209, 399, 231, 275, 180]

# Also check if all original prices are in adjusted prices
df
```

```python
# 6. What % of the rooms are private, and what % of the rooms are shared.  

# Normalize, multiply by 100 and then used a lambda function to round to 3 decimale place at least
room_percentages = (df["room_type"].value_counts(normalize=True) * 100).apply(lambda x : f'{x:.3}%')

# Retrieve the percentages for private and shared room
room_percentages.get(['Private room', 'Shared room'])
# Private room = 45.5%
# Shared room = 2.12%
```

# Exercise 3 - Grouping

1. Using `groupby`, count how many listings are in each neighbourhood_group.


2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 


3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 


4. Using `groupby`, find the median price for each room type in each neighbourhood_group.


5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices for each room type in each neighbourhood_group.

```python
# 1. Using `groupby`, count how many listings are in each neighbourhood_group.

total_group_listings = df.groupby('neighbourhood_group')['id'].count()
total_group_listings
```

```python
# 2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 

df.groupby('neighbourhood_group')['price'].mean().reset_index()
```

```python
# 2.5. Using `groupby`, find the mean price for each room_type. 

df.groupby('room_type')['price'].mean()
```

```python
# 3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 

df.groupby('neighbourhood_group')['price'].agg( ['min', 'max'] )
```

```python
# 4. Using `groupby`, find the mean price for each room_type in each neighbourhood_group.

df.groupby(['neighbourhood_group', 'room_type'])['price'].mean()
```

```python
# 5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices 
# for each room type in each neighbourhood_group.

df.groupby(['neighbourhood_group', 'room_type'])['price'].agg(['count', 'min', 'max', 'median', 'std'])

```

```python
# 1. Load the `prices.csv` and the `n_listings.csv`
df_prices = pd.read_csv('data/prices.csv')
df_prices

df_n_listings = pd.read_csv('data/n_listings.csv', sep=';')
df_n_listings
```

```python
# 2. Do join that keeps all the records for each table.


# for how we dont use left or right but outer gives us all
pd.merge(df_prices, df_n_listings, how='outer')
```

```python
# 3. Save your joined csv as joined.csv.

joined_df = pd.merge(df_prices, df_n_listings, how='outer')

joined_df.to_csv('joined.csv', index=False)

# 4.Load your saved table and see if it looks the same or different that the DataFrame you used to create it

df_loaded_joined = pd.read_csv('joined.csv')
df_loaded_joined
```
